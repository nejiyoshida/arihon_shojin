## 尺取り法
i~jまでの和　＝　0~jまでの和 - 0~i-1までの和  
みたいにするやつ　　
使ってたけど名前は知らなかった

```python

"""
POJ 3061

尺取り＋二分探索
"""

from bisect import bisect_left



N = int(input())
S = int(input())
A = list(map(int, input().split()))

cum = [0] * N
cum[0] = A[0]

for i in range(1,N):
    cum[i] = cum[i-1] + A[i]


if cum[-1] < S:
    print(0)
    exit()


ans = N
i = 0
while cum[i] + S <= cum[-1]:
    # i+1~jの和がS以上となるjを探す
    # cum[j] - cum[i] >= S -> cum[j] >= cum[i] + S となる最小のjを探す
    j = bisect_left(cum, cum[i]+S)
    ans = min(ans, j-i)
    i += 1

print(ans)


"""
10
15
5 1 3 5 10 7 4 9 2 8
2


5
11
1 2 3 4 5
3
"""
```

```python
"""

別解

ｊを増やしていってSを超えるときは何要素分の和か調べてそのあと先頭を一つ捨てる、を合計がSを超えなくなるまで繰り返す
"""

N = int(input())
S = int(input())
A = list(map(int, input().split()))


ans = N+1

i = j = 0
total = 0
while True:
    while j <= N-1 and total < S:

        total += A[j]
        j += 1
    # 可能な限り足してもSを超えないなら終わり
    if total < S:
        break
    ans = min(ans, j-i)
    # Sを超えているので末尾分引く
    total -= A[i]
    i += 1

if ans > N:
    print(0)
else:
    print(ans)
    

```

```python

"""
AtCoder Beginner Contest 032 C


例題の掛け算バージョン
１をかけても累積積は変わらないことなどに注意
"""


N,K = map(int, input().split())

S = [0] * N
for i in range(N):
    s = int(input())
    S[i] = s

if 0 in S:
    print(N)    
    exit()


if min(S) > K:
    print(0)
    exit()

if N == 1:
    print(1 if S[0] <= K else 0)

i = j = 0
total = 1
ans = 1
for i in range(N):
    while j <= N-1 and total*S[j] <= K:
        total *= S[j]
        j += 1
    
    ans = max(ans, j-i)
    total //= S[i]


print(ans)
```

```python

"""
AtCoder Regular Contest 098 D


XORの問題だけど珍しく桁ごとに見なくてよい
尺取り法はiとjの窓をずらす条件を取り違えなければそんなに難しくなくて便利
"""


N = int(input())
A = list(map(int, input().split()))


ex = 0


ans = 0
j = 0
ex = 0
for i in range(N):
    while j <= N-1 and ex + A[j] == ex ^ A[j]:
        ex += A[j]
        j += 1

    ans += j-i  
    ex -= A[i]

print(ans)

```

```python
"""
POJ 3320

s~tページで事柄全てをカバー
→　s+1からであれば、　t'>=tまでは最低でも読む必要ある　

全ての事柄が一回以上出てくるよう読みたい
先頭のsを読まなくした場合に、tまでの範囲でsに書かれている事柄の個数が0になるならそれが出てくるまでtを進める


"""
from collections import defaultdict

P = int(input())
A = list(map(int, input().split()))

contents = len(set(A))
ans = P
i = 0
j = 0
# その範囲にある事柄の数
cnt = 0
d = defaultdict(int)
for i in range(P):
    while j <= P-1 and cnt < contents:
        if d[A[j]] == 0:
            cnt += 1
        d[A[j]] += 1
        j += 1
    
    # 最後まで見ても事柄を全て見切れないなら終わり
    if cnt < contents:
        break

    ans = min(ans, j - i)
    d[A[i]] -= 1

    if d[A[i]] == 0:
        cnt -= 1

print(ans)
```

```python
"""
AtCoder Regular Contest 022 B

jについて、同じ味が出てくるまで進める（どの味がすでに出てきているか残しておく）
同じ味が出てきたら、被りのない範囲と現時点の最長部分の大きいほうを答えにする。
先頭の味を切り捨てて、初めに戻る

を繰り返す。
"""

from collections import defaultdict

N = int(input())
A = list(map(int, input().split()))


ans = 0
j = 0
d = defaultdict(int)

for i in range(N):

    while j <= N-1 and d[A[j]] == 0:
        d[A[j]] += 1
        j += 1

    
    ans = max(ans, j-i)
    d[A[i]] -= 1

print(ans)
```

```python
"""
JOI 2013 本選 C バームクーヘン　
https://atcoder.jp/contests/joi2014ho/tasks/joi2014ho3

二分探索を利用
ある切り込みから尺取りでmidを超える、最も近い切れ込み部分で切り分ける、を２回やると、ピースが３つできる。
この中で一番小さい物がmid以上なら最小がmidは作れるので、lo=mid（これは作れる）にする。
mid未満であればそれはできないのでhi=mid(このhi作れないから答えに選べない)
だいぶ難しい
"""

from bisect import bisect_left

N = int(input())
A = [int(input()) for _ in range(N)]
total = sum(A)


cum = [0]
for i in range(N):
    cum.append(cum[-1]+A[i])

hi = total
lo = 0

# 最小のものがmidを越せるか確認
# [li, hi)に求める値がある。hiは選択できない。最終的なloが答え
while hi - lo > 1:
    mid = (hi+lo)//2
    f = False
    for i in range(len(cum)):
        # 一つ目の選んだ切り込みからmid離れたところえお切ろうとすると足りない場合
        if cum[i] + mid > total:
            continue
        j = bisect_left(cum, cum[i]+mid)
        p1 = cum[j] - cum[i]
        # ２つ目の選んだ切り込みからmid離れたところえお切ろうとすると足りない場合
        if cum[j] + mid > total:
            continue
        k = bisect_left(cum, cum[j] + mid)
        p2 = cum[k] - cum[j]
        # ３つのうち一番小さいのがmidを超えていればOK
        if min(p1, p2, total-p1-p2) >= mid:
            f = True
            break
    
    if f:
        lo = mid
    else:
        hi = mid
    #mid = (hi+lo)//2

print(lo)
```

```python
"""
POJ 3276

難しい。
・Kを全探索
・どれを起点に反転させるか
・ある牛がその時点で何回反転しているか
を網羅的に見ていくと、O(N^3)となり、間に合わない。

i番目の牛を起点に反転させると、その影響はi+K-1番目にまで及ぶので、
i-K+1 ~ i-1番目までがひっくり返った回数の総和を変数で持っておいて、
i-> i+1とする際に、i-K+1での操作の影響がi+1に及ばないようにとりのぞいてやると、O(N^2)にでき、間に合う
"""


N = int(input())
S = input()

bf = [0] * N
for i in range(N):
    if S[i] == "F":
        bf[i] = 0
    else:
        bf[i] = 1

def solve(k):
    cnt = 0
    # 区間[i, i+K-1]を反転させた回数
    f = [0] * N
    # 
    total = 0
    for i in range(0, N-k+1):
        # iからi+k-1の区間で、先頭の牛が後ろを向いている場合
        # total はi番目の牛が、その時点でひっくり返された回数
        if (bf[i] + total) % 2 != 0:
            # その区間の牛の向きを変えるので＋１
            cnt += 1
            # i番目の向きを変えた
            f[i] = 1
        
        total += f[i]
        # i-K+1の牛をもしひっくり返していたら、iもひっくり返っている
        # その牛より左にk-1頭以上の牛がいたら、i-k+1牛をひっくり返していれば、その操作を除く
        # というのも、i-k+1の操作の影響は、iまでにしか及ばないので、i++する前にその影響を除く必要があるため。
        if i-k+1 >= 0:
            total -= f[i-k+1]
        
    # 残りが前を向いているか調べる
    # 前からk頭は区間の先頭になりうるので、どちらにしろ前向きになる
    for i in range(N-k+1, N):
        if (bf[i] + total) % 2 != 0:
            return -1
        # 上と同じ処理。
        # i-k+1での操作が、iより右に及ばないように調整
        if i-k+1 >= 0:
            total -= f[i-k+1]
        
    return cnt

ans_m = N+1
ans_k = N+1
for k in range(1,N+1):
    m = solve(k)
    if m >= 0 and ans_m > m:
        ans_m = m
        ans_k = k
print(ans_k, ans_m)

"""
7
BBFBFBB

3 3
"""
```

```python
"""
AtCoder Regular Contest 064

区間ごとに決めていく貪欲
・前から順番に減らしていく。直前の箱と合計してxを超えるなら、超える分だけ減らす。
・A[i]+A[i+1]でどちらを減らせばいいか考える場合、A[i+1]が小さければ、次のA[i+1]+A[i+2]を見ていくときにもいい影響を与えられるので、A[i+1]の箱から食べるほうがいい
　
"""


N,x = map(int, input().split())
A = list(map(int, input().split()))


ans = 0
for i in range(N):
    if i == 0:
        if A[i] > x:
            ans += A[i] - x
            A[i] = x

    else:
        if A[i-1] + A[i] > x:
            ans += A[i-1] + A[i] - x
            A[i] = x - A[i-1]

print(ans)
```

```python
"""
AtCoder Regular Contest 088　D

S = 000011111111100
で考えてみる
i=3の0と=4の1について、文字が異なるので、その左か右のどちらかで書き換えが必要
左で書き換えると0000が1111になり、そのあとそうさを続けてもK=4となる。
右で書き換えると、0000000000000011となり、続けていくとK=11となる。

文字がiとi+1で異なるところで、左右の部分文字列の長い方を書き換えると、その時点でKをなるべく減らさなくてすむ。
iを先頭から末尾まで見ていき、Kとして選ばれたもののなかで、一番小さいものが答え。
"""

S = input()

N = len(S)
# 文字列がすべて同じ文字ならK=Nになるので
K = N
for i in range(N-1):
    # iとその右の文字が異なる場合、長い方を書き換えるとKをなるべく大きく保てる
    # Kの減る幅をなるべく小さように選んでいくイメージ。
    if S[i] != S[i+1]:
        tmp = max(i+1, N - (i+1))
        K = min(K, tmp)

print(K)
```
