## 尺取り法
i~jまでの和　＝　0~jまでの和 - 0~i-1までの和  
みたいにするやつ

```python

"""
POJ 3061

尺取り＋二分探索
"""

from bisect import bisect_left



N = int(input())
S = int(input())
A = list(map(int, input().split()))

cum = [0] * N
cum[0] = A[0]

for i in range(1,N):
    cum[i] = cum[i-1] + A[i]


if cum[-1] < S:
    print(0)
    exit()


ans = N
i = 0
while cum[i] + S <= cum[-1]:
    # i+1~jの和がS以上となるjを探す
    # cum[j] - cum[i] >= S -> cum[j] >= cum[i] + S となる最小のjを探す
    j = bisect_left(cum, cum[i]+S)
    ans = min(ans, j-i)
    i += 1

print(ans)


"""
10
15
5 1 3 5 10 7 4 9 2 8
2


5
11
1 2 3 4 5
3
"""
```

```python
"""
POJ 3320

s~tページで事柄全てをカバー
→　s+1からであれば、　t'>=tまでは最低でも読む必要ある　

全ての事柄が一回以上出てくるよう読みたい
先頭のsを読まなくした場合に、tまでの範囲でsに書かれている事柄の個数が0になるならそれが出てくるまでtを進める


"""
from collections import defaultdict

P = int(input())
A = list(map(int, input().split()))

contents = len(set(A))
ans = P
i = 0
j = 0
# その範囲にある事柄の数
cnt = 0
d = defaultdict(int)
for i in range(P):
    while j <= P-1 and cnt < contents:
        if d[A[j]] == 0:
            cnt += 1
        d[A[j]] += 1
        j += 1
    
    # 最後まで見ても事柄を全て見切れないなら終わり
    if cnt < contents:
        break

    ans = min(ans, j - i)
    d[A[i]] -= 1

    if d[A[i]] == 0:
        cnt -= 1

print(ans)
```

```python
"""
JOI 2013 本選 C バームクーヘン　
https://atcoder.jp/contests/joi2014ho/tasks/joi2014ho3

二分探索を利用
ある切り込みから尺取りでmidを超える、最も近い切れ込み部分で切り分ける、を２回やると、ピースが３つできる。
この中で一番小さい物がmid以上なら最小がmidは作れるので、lo=mid（これは作れる）にする。
mid未満であればそれはできないのでhi=mid(このhi作れないから答えに選べない)
だいぶ難しい
"""

from bisect import bisect_left

N = int(input())
A = [int(input()) for _ in range(N)]
total = sum(A)


cum = [0]
for i in range(N):
    cum.append(cum[-1]+A[i])

hi = total
lo = 0

# 最小のものがmidを越せるか確認
# [li, hi)に求める値がある。hiは選択できない。最終的なloが答え
while hi - lo > 1:
    mid = (hi+lo)//2
    f = False
    for i in range(len(cum)):
        # 一つ目の選んだ切り込みからmid離れたところえお切ろうとすると足りない場合
        if cum[i] + mid > total:
            continue
        j = bisect_left(cum, cum[i]+mid)
        p1 = cum[j] - cum[i]
        # ２つ目の選んだ切り込みからmid離れたところえお切ろうとすると足りない場合
        if cum[j] + mid > total:
            continue
        k = bisect_left(cum, cum[j] + mid)
        p2 = cum[k] - cum[j]
        # ３つのうち一番小さいのがmidを超えていればOK
        if min(p1, p2, total-p1-p2) >= mid:
            f = True
            break
    
    if f:
        lo = mid
    else:
        hi = mid
    #mid = (hi+lo)//2

print(lo)
```

```python
"""
POJ 3276

難しい。
・Kを全探索
・どれを起点に反転させるか
・ある牛がその時点で何回反転しているか
を網羅的に見ていくと、O(N^3)となり、間に合わない。

i番目の牛を起点に反転させると、その影響はi+K-1番目にまで及ぶので、
i-K+1 ~ i-1番目までがひっくり返った回数の総和を変数で持っておいて、
i-> i+1とする際に、i-K+1での操作の影響がi+1に及ばないようにとりのぞいてやると、O(N^2)にでき、間に合う
"""


N = int(input())
S = input()

bf = [0] * N
for i in range(N):
    if S[i] == "F":
        bf[i] = 0
    else:
        bf[i] = 1

def solve(k):
    cnt = 0
    # 区間[i, i+K-1]を反転させた回数
    f = [0] * N
    # 
    total = 0
    for i in range(0, N-k+1):
        # iからi+k-1の区間で、先頭の牛が後ろを向いている場合
        # total はi番目の牛が、その時点でひっくり返された回数
        if (bf[i] + total) % 2 != 0:
            # その区間の牛の向きを変えるので＋１
            cnt += 1
            # i番目の向きを変えた
            f[i] = 1
        
        total += f[i]
        # i-K+1の牛をもしひっくり返していたら、iもひっくり返っている
        # その牛より左にk-1頭以上の牛がいたら、i-k+1牛をひっくり返していれば、その操作を除く
        # というのも、i-k+1の操作の影響は、iまでにしか及ばないので、i++する前にその影響を除く必要があるため。
        if i-k+1 >= 0:
            total -= f[i-k+1]
        
    # 残りが前を向いているか調べる
    # 前からk頭は区間の先頭になりうるので、どちらにしろ前向きになる
    for i in range(N-k+1, N):
        if (bf[i] + total) % 2 != 0:
            return -1
        # 上と同じ処理。
        # i-k+1での操作が、iより右に及ばないように調整
        if i-k+1 >= 0:
            total -= f[i-k+1]
        
    return cnt

ans_m = N+1
ans_k = N+1
for k in range(1,N+1):
    m = solve(k)
    if m >= 0 and ans_m > m:
        ans_m = m
        ans_k = k
print(ans_k, ans_m)

"""
7
BBFBFBB

3 3
"""
```

```python
"""
AtCoder Regular Contest 064

区間ごとに決めていく貪欲
・前から順番に減らしていく。直前の箱と合計してxを超えるなら、超える分だけ減らす。
・A[i]+A[i+1]でどちらを減らせばいいか考える場合、A[i+1]が小さければ、次のA[i+1]+A[i+2]を見ていくときにもいい影響を与えられるので、A[i+1]の箱から食べるほうがいい
　
"""


N,x = map(int, input().split())
A = list(map(int, input().split()))


ans = 0
for i in range(N):
    if i == 0:
        if A[i] > x:
            ans += A[i] - x
            A[i] = x

    else:
        if A[i-1] + A[i] > x:
            ans += A[i-1] + A[i] - x
            A[i] = x - A[i-1]

print(ans)
```

```python
"""
AtCoder Regular Contest 088　D

S = 000011111111100
で考えてみる
i=3の0と=4の1について、文字が異なるので、その左か右のどちらかで書き換えが必要
左で書き換えると0000が1111になり、そのあとそうさを続けてもK=4となる。
右で書き換えると、0000000000000011となり、続けていくとK=11となる。

文字がiとi+1で異なるところで、左右の部分文字列の長い方を書き換えると、その時点でKをなるべく減らさなくてすむ。
iを先頭から末尾まで見ていき、Kとして選ばれたもののなかで、一番小さいものが答え。
"""

S = input()

N = len(S)
# 文字列がすべて同じ文字ならK=Nになるので
K = N
for i in range(N-1):
    # iとその右の文字が異なる場合、長い方を書き換えるとKをなるべく大きく保てる
    # Kの減る幅をなるべく小さように選んでいくイメージ。
    if S[i] != S[i+1]:
        tmp = max(i+1, N - (i+1))
        K = min(K, tmp)

print(K)
```

```python

"""
第７回日本情報オリンピック 予選（オンライン） E - おせんべい


全てのマスを裏にはできないので、いい感じに縦横ひっくり返すものを選んで全体を最大化する
ひっくり返し方は2**(R*C)通りあるので全探索は無理
あるおせんべいについて考えるとき、縦か横のひっくり返し方が決まっていれば、もう片方をひっくり返すべきかどうかが決まる
なので、数の小さいRについて、ひっくり返し方を決め打って、その時にそれぞれの列をひっくり返すかどうか見ていく
決め打ち方は全探索できて、2**R通り

"""

R,C = map(int, input().split())
A = [tuple(map(int, input().split())) for _ in range(R)]

ans = 0
# bitの1が立っている行をひっくり返す
for bit in range(1 << R):
    curr = 0
    # それぞれの列について、多いほうを裏として採用する
    for i in range(C):
        cnt_rev = 0
        for j in range(R):
            if (not bit & (1 << j) and A[j][i] == 0) or (bit & (1 << j) and A[j][i] != 0):
                cnt_rev += 1
        curr += max(cnt_rev, R - cnt_rev)
    
    ans = max(ans, curr)

print(ans)

```

```python
"""
第７回日本情報オリンピック 本選（オンライン）　　C - ダーツ

ダーツを４本投げて合計点がMを超えない最大をさがす
こういう最大を探すのは二分探索が強め。
また、４本投げたパターンを全探索はしんどいので、２本投げたときの得点を全探索してそれを２セットみたいな。
また、投げない＝０にあたる、と考える
"""

from bisect import bisect_right

N,M = map(int, input().split())
P = [int(input()) for _ in range(N)]


P = [0] + P
d2 = set()
for i in range(N+1):
    for j in range(N+1):
        if P[i] + P[j] <= M:
            d2.add(P[i] + P[j])
d2 = list(d2)
d2.sort()
ans = 0

for n in d2:
    rest = M - n
    idx = bisect_right(d2, rest)
    if d2[idx - 1] + n <= M:
        ans = max(ans, d2[idx - 1] + n)

print(ans)
```

```python
"""
Huge Knapsack Problem
http://judge.u-aizu.ac.jp/onlinejudge/description.jsp?id=DPL_1_H&lang=jp

半分全列挙で前半と後半の選び方を全て作っておいて、それらの組み合わせで調べる
pythonでこれだとTLEだが、ロジックはあっている
"""

from bisect import bisect_right


def main():

    N,W = map(int, input().split())
    wv = []
    for _ in range(N):
        v,w = map(int, input().split())
        wv.append([w,v])


    half = N // 2
    pairs = [[0,0]] # 前半から何も選ばない場合も入れておく
    # bit全探索の要領で半分全列挙
    for bit in range(1 << half):
        w = 0
        v = 0
        for i in range(half):
            # bitのi番目が立っていれば、前半のi番目をナップザックに入れるパターン
            if bit & (1 << i):
                w += wv[i][0]
                v += wv[i][1]
        

        if w <= W:    
            pairs.append([w,v])

    pairs.sort()
    # 軽い組の方が価値があるなら、必要ない組をつぶす
    # これをしないと後でバグる（重いけど価値のない組が選ばれる）
    for i in range(len(pairs)-1):
        if pairs[i][1] > pairs[i+1][1]:
            pairs[i+1][1] = pairs[i][1]


    ans = 0
    pairs2 = [[0,0]]
    # 後半の全列挙
    for bit in range(1 << (N - half)):
        w = 0
        v = 0
        for i in range(N - half):
            if bit & (1 << i):
                w += wv[half + i][0]
                v += wv[half + i][1]
        
        # 制限を超えない場合
        if w <= W:    
            pairs2.append([w,v])

    pairs2.sort()
    # 軽い組の方が価値があるなら、必要ない組をつぶす
    # これをしないと後でバグる（重いけど価値のない組が選ばれる）
    for i in range(len(pairs2)-1):
        if pairs2[i][1] > pairs2[i+1][1]:
            pairs2[i+1][1] = pairs2[i][1]

    # 後半から何も選ばない場合

    for w,v in pairs2:
        idx = bisect_right(pairs, [W-w, float("inf")])
        ans = max(ans, pairs[idx-1][1] + v)

    print(ans)


if __name__ == "__main__":
    main()
```

```python
"""
Huge Knapsack Problem
http://judge.u-aizu.ac.jp/onlinejudge/description.jsp?id=DPL_1_H&lang=jp

半分全列挙で前半と後半の選び方を全て作っておいて、それらの組み合わせで調べる
こっちはbit演算なしでやるようにしたので間に合う

"""

from bisect import bisect_right


def main():

    N,W = map(int, input().split())
    wv = []
    for _ in range(N):
        v,w = map(int, input().split())
        wv.append([w,v])

    d1 = [[0,0]]
    d2 = [[0,0]]
    half = N // 2

    for i in range(half):
        w,v = wv[i]
        size = len(d1)
        # i-1までに作っている組について、iを足すパターンを加える
        for j in range(size):
            d1.append([d1[j][0] + w, d1[j][1] + v])

    for i in range(N - half):
        w,v = wv[half + i]
        size = len(d2)
        for j in range(size):
            d2.append([d2[j][0] + w, d2[j][1] + v])

    d1.sort()
    d2.sort()

    # i+1番目はiより重いが、iより価値が低いなら、i+1の重さまででiの価値ができるやで、に塗り替える
    for i in range(len(d2) - 1):
        if d2[i+1][1] < d2[i][1]:
            d2[i+1][1] = d2[i][1]

    ans = 0
    INF = float("inf")
    for w,v in d1:
        if w > W:
            continue

        rest = W - w
        idx = bisect_right(d2, [rest, INF])
        ans = max(ans, d2[idx - 1][1] + v)

    print(ans)

if __name__ == "__main__":
    main()
```

```python
"""

AtCoder Grand Contest 026  C - String Coloring



超ムズイ
N <= 18なので、何らかbit操作しそう。
len(S) = 2N　も謎だが、Sが常に偶数長さになるので、半分にしてほしそうな感じ
nCr(36,18)はすごい数字になるので無理

解説見ると、
文字列を前後Nで分けて、夫々を赤青に塗りわけたものが、反対側の逆の色と同じになればよいのこと。
また、青は逆から読むので、後半N文字をひっくり返してしまうと、同じ色が同じ文字になればいいらしい。
で、青と赤はN文字について半分全探索。

これは出てこない......
"""

from collections import defaultdict


def main():
        
    N = int(input())
    S = input()

    X = S[:N]
    Y = S[N:]
    Y = Y[::-1]

    dx = defaultdict(int)
    dy = defaultdict(int)

    for bit in range(1 << N):
        red = ""
        blue = ""
        for i in range(N):
            if bit & (1 << i):
                red += X[i]
            else:
                blue += X[i]

        dx[(red, blue)] += 1

    for bit in range(1 << N):
        red = ""
        blue = ""
        for i in range(N):
            if bit & (1 << i):
                red += Y[i]
            else:
                blue += Y[i]

        dy[(red, blue)] += 1

    # 前後が同じになる組がいくつできるか。
    ans = 0
    for k,v in dx.items():
        ans += dy[k] * v

    print(ans)


if __name__ == "__main__":
    main()
```

```python
"""

AtCoder Regular Contest 017  C - 無駄なものが嫌いな人

N <= 40ぐらいまでで、愚直にやると2^N　になる問題は半分全列挙をやってみるといいかも。
今回は品物を半分にわけて、ぞれぞれで作りうる大きさとその個数を用意し、最後に大きさがXになる組を突き合わせ。

"""



from collections import Counter


N,X = map(int, input().split())
W = [int(input()) for _ in range(N)]


half = N // 2
A = W[:half]
B = W[half:]

bag_A = [0]
bag_B = [0]

for a in A:
    size_a = len(bag_A)
    for i in range(size_a):
        w = bag_A[i]
        if a + w <= X:
            bag_A.append(a + w)

for b in B:
    size_b = len(bag_B)
    for i in range(size_b):
        w = bag_B[i]
        if b + w <= X:
            bag_B.append(b + w)

ans = 0
CA = Counter(bag_A)
CB = Counter(bag_B)

for k,v in CB.items():
    if X - k in CA:
        ans += v * CA[X - k]

print(ans)
```
