## 尺取り法
i~jまでの和　＝　0~jまでの和 - 0~i-1までの和  
みたいにするやつ　　
使ってたけど名前は知らなかった

```python

"""
POJ 3061

尺取り＋二分探索
"""

from bisect import bisect_left



N = int(input())
S = int(input())
A = list(map(int, input().split()))

cum = [0] * N
cum[0] = A[0]

for i in range(1,N):
    cum[i] = cum[i-1] + A[i]


if cum[-1] < S:
    print(0)
    exit()


ans = N
i = 0
while cum[i] + S <= cum[-1]:
    # i+1~jの和がS以上となるjを探す
    # cum[j] - cum[i] >= S -> cum[j] >= cum[i] + S となる最小のjを探す
    j = bisect_left(cum, cum[i]+S)
    ans = min(ans, j-i)
    i += 1

print(ans)


"""
10
15
5 1 3 5 10 7 4 9 2 8
2


5
11
1 2 3 4 5
3
"""
```

```python
"""

別解

ｊを増やしていってSを超えるときは何要素分の和か調べてそのあと先頭を一つ捨てる、を合計がSを超えなくなるまで繰り返す
"""

N = int(input())
S = int(input())
A = list(map(int, input().split()))


ans = N+1

i = j = 0
total = 0
while True:
    while j <= N-1 and total < S:

        total += A[j]
        j += 1
    # 可能な限り足してもSを超えないなら終わり
    if total < S:
        break
    ans = min(ans, j-i)
    # Sを超えているので末尾分引く
    total -= A[i]
    i += 1

if ans > N:
    print(0)
else:
    print(ans)
    

```
