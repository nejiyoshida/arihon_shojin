## 尺取り法
i~jまでの和　＝　0~jまでの和 - 0~i-1までの和  
みたいにするやつ　　
使ってたけど名前は知らなかった

```python

"""
POJ 3061

尺取り＋二分探索
"""

from bisect import bisect_left



N = int(input())
S = int(input())
A = list(map(int, input().split()))

cum = [0] * N
cum[0] = A[0]

for i in range(1,N):
    cum[i] = cum[i-1] + A[i]


if cum[-1] < S:
    print(0)
    exit()


ans = N
i = 0
while cum[i] + S <= cum[-1]:
    # i+1~jの和がS以上となるjを探す
    # cum[j] - cum[i] >= S -> cum[j] >= cum[i] + S となる最小のjを探す
    j = bisect_left(cum, cum[i]+S)
    ans = min(ans, j-i)
    i += 1

print(ans)


"""
10
15
5 1 3 5 10 7 4 9 2 8
2


5
11
1 2 3 4 5
3
"""
```

```python
"""

別解

ｊを増やしていってSを超えるときは何要素分の和か調べてそのあと先頭を一つ捨てる、を合計がSを超えなくなるまで繰り返す
"""

N = int(input())
S = int(input())
A = list(map(int, input().split()))


ans = N+1

i = j = 0
total = 0
while True:
    while j <= N-1 and total < S:

        total += A[j]
        j += 1
    # 可能な限り足してもSを超えないなら終わり
    if total < S:
        break
    ans = min(ans, j-i)
    # Sを超えているので末尾分引く
    total -= A[i]
    i += 1

if ans > N:
    print(0)
else:
    print(ans)
    

```

```python

"""
AtCoder Beginner Contest 032 C


例題の掛け算バージョン
１をかけても累積積は変わらないことなどに注意
"""


N,K = map(int, input().split())

S = [0] * N
for i in range(N):
    s = int(input())
    S[i] = s

if 0 in S:
    print(N)    
    exit()


if min(S) > K:
    print(0)
    exit()

if N == 1:
    print(1 if S[0] <= K else 0)

i = j = 0
total = 1
ans = 1
for i in range(N):
    while j <= N-1 and total*S[j] <= K:
        total *= S[j]
        j += 1
    
    ans = max(ans, j-i)
    total //= S[i]


print(ans)
```

```python
# ちゃんとしたしゃくとりの解法



N = int(input())
A = list(map(int, input().split()))

j = 1
ans = 0

for i in range(N):
    # jとiが同じ位置にきたら範囲を取れないのでずらす
    if j == i:
        j += 1
    while j <= N - 1 and A[j] > A[j-1]:
        j += 1

    ans += j - i

print(ans)
```

```python

"""
AtCoder Regular Contest 098 D


XORの問題だけど珍しく桁ごとに見なくてよい
尺取り法はiとjの窓をずらす条件を取り違えなければそんなに難しくなくて便利
"""


N = int(input())
A = list(map(int, input().split()))


ex = 0


ans = 0
j = 0
ex = 0
for i in range(N):
    while j <= N-1 and ex + A[j] == ex ^ A[j]:
        ex += A[j]
        j += 1

    ans += j-i  
    ex -= A[i]

print(ans)

```

```python
"""
POJ 3320

s~tページで事柄全てをカバー
→　s+1からであれば、　t'>=tまでは最低でも読む必要ある　

全ての事柄が一回以上出てくるよう読みたい
先頭のsを読まなくした場合に、tまでの範囲でsに書かれている事柄の個数が0になるならそれが出てくるまでtを進める


"""
from collections import defaultdict

P = int(input())
A = list(map(int, input().split()))

contents = len(set(A))
ans = P
i = 0
j = 0
# その範囲にある事柄の数
cnt = 0
d = defaultdict(int)
for i in range(P):
    while j <= P-1 and cnt < contents:
        if d[A[j]] == 0:
            cnt += 1
        d[A[j]] += 1
        j += 1
    
    # 最後まで見ても事柄を全て見切れないなら終わり
    if cnt < contents:
        break

    ans = min(ans, j - i)
    d[A[i]] -= 1

    if d[A[i]] == 0:
        cnt -= 1

print(ans)
```

```python
"""
AtCoder Regular Contest 022 B

jについて、同じ味が出てくるまで進める（どの味がすでに出てきているか残しておく）
同じ味が出てきたら、被りのない範囲と現時点の最長部分の大きいほうを答えにする。
先頭の味を切り捨てて、初めに戻る

を繰り返す。
"""

from collections import defaultdict

N = int(input())
A = list(map(int, input().split()))


ans = 0
j = 0
d = defaultdict(int)

for i in range(N):

    while j <= N-1 and d[A[j]] == 0:
        d[A[j]] += 1
        j += 1

    
    ans = max(ans, j-i)
    d[A[i]] -= 1

print(ans)
```

```python
"""
JOI 2013 本選 C バームクーヘン　
https://atcoder.jp/contests/joi2014ho/tasks/joi2014ho3

二分探索を利用
ある切り込みから尺取りでmidを超える、最も近い切れ込み部分で切り分ける、を２回やると、ピースが３つできる。
この中で一番小さい物がmid以上なら最小がmidは作れるので、lo=mid（これは作れる）にする。
mid未満であればそれはできないのでhi=mid(このhi作れないから答えに選べない)
だいぶ難しい
"""

from bisect import bisect_left

N = int(input())
A = [int(input()) for _ in range(N)]
total = sum(A)


cum = [0]
for i in range(N):
    cum.append(cum[-1]+A[i])

hi = total
lo = 0

# 最小のものがmidを越せるか確認
# [li, hi)に求める値がある。hiは選択できない。最終的なloが答え
while hi - lo > 1:
    mid = (hi+lo)//2
    f = False
    for i in range(len(cum)):
        # 一つ目の選んだ切り込みからmid離れたところえお切ろうとすると足りない場合
        if cum[i] + mid > total:
            continue
        j = bisect_left(cum, cum[i]+mid)
        p1 = cum[j] - cum[i]
        # ２つ目の選んだ切り込みからmid離れたところえお切ろうとすると足りない場合
        if cum[j] + mid > total:
            continue
        k = bisect_left(cum, cum[j] + mid)
        p2 = cum[k] - cum[j]
        # ３つのうち一番小さいのがmidを超えていればOK
        if min(p1, p2, total-p1-p2) >= mid:
            f = True
            break
    
    if f:
        lo = mid
    else:
        hi = mid
    #mid = (hi+lo)//2

print(lo)
```
