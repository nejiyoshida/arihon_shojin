```python
"""
AtCoder Regular Contest 040 B



最小値→二分探索的雰囲気を感じるがどうだろうか。Nが小さいので二分探索使わない雰囲気の方が強いが
・塗るを選択すると足元含んで前方のRマスを塗る
・移動は単に前に一マス進む

左から始めて、
・R分一回塗ったら全マス塗りきれる場合、一番右側にある塗っていないマスを塗れるギリギリの位置まで進んで塗る
→R分塗ったら終わり：残りの塗っていないマスの数がR個以下で、それがRの長さの中に存在

・R分一回塗っても塗り残しがある場合、一番左側にある塗っていないマスまで進んで塗る
→R分塗っても終わらない：R個より多く塗り残しがあるOR、Rの長さのなかに収まっていない（散らばってる）

なので、
・一番左の塗ってない奴の位置
・一番右の塗ってない奴の位置
を使って一ステップずつ調べていけばいい

"""


N,R = map(int, input().split())

S = list(input())

# 塗っていないマスの中で一番最後にあるマスのidx
last_idx = -1

for i in reversed(range(N)):
    #一番最後を記録
    if S[i] == ".":
        last_idx = i
        break

#　全部見ても最初から塗ってないマスがないときは操作なしで終わり
if last_idx == -1:
    print(0)
    exit()



ans = 0
curr = 0
next_idx = 0
# 最後を塗り終わるまでループ
while S[last_idx] != "o":
    # 今いる位置を含んで前方で塗っていないマスで一番近い部分を探す
    for i,c in enumerate(S[curr:]):
        if c == ".":
            next_idx = i
            break


    # 一番左にある塗ってないマスと、一番右にある塗ってないマスがR範囲にあれば、同時に塗って終わり
    # その場合は右端のがギリギリ塗れる位置まで移動して塗る
    # Rの範囲になければ、左の塗ってないマスの上で塗る

    # 一気に最後まで塗りきれる場合
    if last_idx - (curr + next_idx) <= R-1:
        # last_idxからR-1分左の位置までの範囲にいる場合、その場で塗って終わり
        if last_idx - (R-1) <= curr <= last_idx:
            ans += 1
        # 最後まで塗りきるのに移動する必要がある場合
        else:
            ans += (last_idx - (R-1)) - curr + 1
        break


    # 一回の塗りでは最後まで塗りきれないので、左の塗ってないマスの上で塗る
    else:
        # 進んだところからRの範囲を塗る
        for i in range(R):
            S[curr + next_idx + i] = "o"
        ans += next_idx + 1
        curr += next_idx

print(ans)
```
