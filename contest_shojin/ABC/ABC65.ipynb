```python
"""
AtCoder Beginner Contest 065 C

10**9+7で割った余りを求める問題は、数が大きくなるので答えでなく計算過程でmodをとっておく。
今回は階乗計算の時にmodをとる
問題自体は高校数学の範囲
"""

N,M = map(int, input().split())
MOD = 10**9+7

if abs(N - M) > 1:
    print(0)
    exit()

factorial = [1] * (max(N,M)+1)
for i in range(1,max(N,M)+1):
    factorial[i] = (factorial[i-1]*i) % MOD

if abs(N - M) == 0:
    ans = ((factorial[N]**2) * 2) % MOD
else:
    ans = (factorial[N] * factorial[M]) % MOD
print(ans)
```

```python
"""
AtCoder Beginner Contest 065 D

全域木にするのに必要な最小コストは？という問題
x座標が小さい順に並べて隣り合うものを辺にしたやつをheapqに追加
y座標が小さい順に並べて隣り合うものを辺にしたやつをheapqに追加
heapqからコストが小さい順に辺を取り出して見ていく。ある辺を採用するかしないか考える場合に、ある辺を見たときに既に両方の町が繋がっている場合はその辺は選ばない。
繋がって居るかどうかはUnionFindが便利
"""
import heapq
from collections import defaultdict

N = int(input())
towns = [[] for _ in range(N)]
for i in range(N):
    x,y = map(int, input().split())
    towns[i] = (i, x, y)

# x,yが小さい順に並び替える
tx = sorted(towns, key=lambda x: x[1])
ty = sorted(towns, key=lambda x: x[2])

# 辺を作ってheapqに突っ込む
q = []
for i in range(1,N):
    heapq.heappush(q, (tx[i][1]-tx[i-1][1], tx[i][0], tx[i-1][0]))
    heapq.heappush(q, (ty[i][2]-ty[i-1][2], ty[i][0], ty[i-1][0]))

class UnionFind:
    def __init__(self, N):
        self.parent = [i for i in range(N)]
        self.rank = [0] * N # 最初はそれぞれのノードが誰ともUnionをつくってないから木の深さが０
        self.count = 0 # unite回数
    
    # 根を返す
    def root(self, a):
        if self.parent[a] == a:
            return a
        else:
            # 親を見に行く
            self.parent[a] = self.root(self.parent[a])
            return self.parent[a]
    
    # 同じUnionに属するかどうか
    def is_same(self, a, b):
        return self.root(a) == self.root(b)
    
    # 同じUnionにする
    def unite(self, a, b):
        ra = self.root(a)
        rb = self.root(b)
        if ra == rb:
            return 
        if self.rank[ra] < self.rank[rb]:
            self.parent[ra] = rb
        else:
            self.parent[rb] = ra
            if self.rank[ra] == self.rank[rb]:
                self.rank[ra] += 1
        self.count += 1

uf = UnionFind(N)
ans = 0
while q:
    cost, a, b = heapq.heappop(q)
    if not uf.is_same(a,b):
        uf.unite(a,b)
        ans += cost

print(ans)
```
