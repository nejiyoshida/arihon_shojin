```python
"""
別解
dequeとか使ってみるのもあり（こっちのほうが多少高速だった）
"""

from fractions import Fraction
from collections import deque
import math

A,B,C,D,E,F = map(int, input().split())

# 作りうる水の量を洗い出す
waters = set()
waters.add(100*A)
if 200*B <= F:
    waters.add(200*B)
q = deque([100*A, 200*B])

while q:
    w = q.popleft()
    # 容器に入れられる量より多いならパス
    if w > F: continue
    for n in [100*A, 200*B]:
        if w + n <= F:
            waters.add(w + n)
            q.append(w + n)

waters = list(waters)
waters.sort()



# 最大濃度
max_conc = -1
# 最大濃度の時の水と砂糖
total = -1
sugar = -1

# それぞれの水量についてなるべく多く砂糖を入れてみる
for w in waters:
    # 砂糖の解けることができる量
    limit = math.floor(w * Fraction(E,100))
    # 全体でFまでしか入らないので、F-wとlimitの小さいほうまでしか追加できない
    limit = min(limit, F - w)
    # Cをi回、Dをj回追加する
    max_sugar = 0
    for i in range(limit // C + 1):
        tmp = C * i + ((limit - C*i) // D) * D
        
        max_sugar = max(max_sugar, tmp)
    if max_conc < Fraction(max_sugar, (w+max_sugar)):
        max_conc = Fraction(max_sugar, (w+max_sugar))
        total = w+max_sugar
        sugar = max_sugar

print(total, sugar)
```
