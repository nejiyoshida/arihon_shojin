```python
"""
二次元累積和


普通の累積和を一番上の行でとって、それより下の行は、その行だけの累積和を変数に持ちつつ、一個上の行の累積和を足していく形
詳細
https://paiza.hatenablog.com/entry/2014/05/28/%E3%82%82%E3%81%97%E5%A5%B3%E5%AD%90%E5%A4%A7%E7%94%9F%E3%83%97%E3%83%AD%E3%82%B0%E3%83%A9%E3%83%9E%E3%81%AB%E3%80%8E%E3%82%A2%E3%83%AB%E3%82%B4%E3%83%AA%E3%82%BA%E3%83%A0%E3%80%8F%E3%82%92%E5%9B%B3
"""

# arr[][] があるとする
def make_2d_accum(arr, h, w):
    accum_2d = [[0 for _ in range(w)] for _ in range(h)]
    accum_2d[0][0] = arr[0][0]
    # 一行目なので、ここは一次元と同じ
    for j in range(1,w):
        accum_2d[0][j] = accum_2d[0][j-1] + arr[0][j]
    # 下の部分の累積和なので二次元
    for i in range(1, h):
        # i行目の累積和
        accum_row_i = 0
        for j in range(w):
            accum_row_i += arr[i][j]
            # 一行上までとi行目を足して２次元
            accum_2d[i][j] = accum_2d[i-1][j] + accum_row_i

    return accum_2d

# (a,b)～(c,d)の部分の累積和。全部同じ(a=b=c=d)だと一マス分
def calc_accum_rectangle(accum_2d, a, b, c, d):
    if a > c or b > d:
        return 0
    elif a == 0 and b == 0:
        return accum_2d[c][d]
    elif a == 0:
        return accum_2d[c][d] - accum_2d[c][b-1]
    elif b == 0:
        return accum_2d[c][d] - accum_2d[a-1][d]
    else:
        return accum_2d[c][d] - accum_2d[a-1][d] - accum_2d[c][b-1] + accum_2d[a-1][b-1]
```

```python
"""
AtCoder Regular Contest 025  B - チョコレート

いかにも２次元累積和をやってくれというような問題。
白と黒の一方を負にしておけば問題ない。
"""

# arr[][] があるとする
def make_2d_accum(arr, h, w):
    accum_2d = [[0 for _ in range(w)] for _ in range(h)]
    accum_2d[0][0] = arr[0][0]
    # 一行目なので、ここは一次元と同じ
    for j in range(1,w):
        accum_2d[0][j] = accum_2d[0][j-1] + arr[0][j]
    # 下の部分の累積和なので二次元
    for i in range(1, h):
        # i行目の累積和
        accum_row_i = 0
        for j in range(w):
            accum_row_i += arr[i][j]
            # 一行上までとi行目を足して２次元
            accum_2d[i][j] = accum_2d[i-1][j] + accum_row_i

    return accum_2d

# (a,b)～(c,d)の部分の累積和。全部同じ(a=b=c=d)だと一マス分
def calc_accum_rectangle(accum_2d, a, b, c, d):
    if a > c or b > d:
        return 0
    elif a == 0 and b == 0:
        return accum_2d[c][d]
    elif a == 0:
        return accum_2d[c][d] - accum_2d[c][b-1]
    elif b == 0:
        return accum_2d[c][d] - accum_2d[a-1][d]
    else:
        # ２項目と３項目の重複部分を足すのを忘れない
        return accum_2d[c][d] - accum_2d[a-1][d] - accum_2d[c][b-1] + accum_2d[a-1][b-1]


H,W = map(int, input().split())
C = [list(map(int, input().split())) for _ in range(H)]


for i in range(H):
    for j in range(W):
        if i % 2  != j % 2:
            C[i][j] *= -1

accum_2d = make_2d_accum(C, H, W)

ans = 0
for b in range(W):
    for a in range(H):
        for d in range(b, W):
            for c in range(a, H):
                # 指定の範囲の二次元累積和が0であれば、その面積を現時点の最大と比較
                if calc_accum_rectangle(accum_2d, a, b, c, d) == 0:
                    ans = max(ans, (c - a + 1)*(d - b + 1))

print(ans)
```

```python
"""
heapqの使い道

ダイクストラ法とかでコストの確定してるコストが小さいノードから探索する～みたいに使うことが多いが、ほかの使い道もある。
たとえば、要素数N（10**5とか）の配列に対して、クエリ(l,r)が与えられるとき、配列の[l:r]にたいして＋１とか（ほかの値でもいいけど）する場合を考える。
愚直にクエリを実行していくと、クエリの数 x 配列サイズ　の計算量となり、膨大な実行時間がかかる。

ここでheapqをうまいこと使うと、計算量を減らせる。
下の問題がその例普通の問題ならTLE
"""
```

```python
"""
AtCoder Beginner Contest 014  C - AtColor


愚直に実行すると、(10**6 + 1) * N　の計算量になり間に合わない
ある濃度について、購入したい人が何人いるか考えるとき、
・事前にaが小さい順に並べておいた購入情報から、今考えている濃度より購入範囲の下限が小さいものを取り出す
・その数の分だけ購入人数をプラスする
で、その濃度の購入人数が分かる。
また、ここまででとりだした購入情報について、次の濃度の購入人数に影響しない（bが次の濃度より小さい）場合もある。
それらをのぞくため、
・上で取り出した購入情報は、上限が小さい順にheapqに積んでおく
・次の濃度について確認する前に、無向になる情報をheapqから捨て、捨てた数だけ購入人数から引いておく
でOK
"""


import heapq


N = int(input())
q = []
for _ in range(N):
    a,b = map(int, input().split())
    heapq.heappush(q, (a,b))


q_effective = []
# 濃度ごとに確認
points = [0] * (10**6 + 1)

curr_point = 0
for i in range(10**6 + 1):
    # 現在見ている濃度以下の購入範囲の下限がある場合は、得点を足す
    while len(q) >= 1 and q[0][0] <= i:
        curr_point += 1
        a,b = heapq.heappop(q)
        heapq.heappush(q_effective, (b, a))

    points[i] = curr_point

    # 次のiに進んだ時に、ある購入範囲が次のiの人気に寄与しなくなる物があるなら除外する
    while len(q_effective) >= 1 and q_effective[0][0] <= i:
        curr_point -= 1
        _,_ = heapq.heappop(q_effective)

print(max(points))
```

```python
"""

AtCoder Regular Contest 038  B - マス目と駒


完全情報ゲームの問題
これは終わりの方から再帰的に開始状態に戻るようにすると解ける。
今回は、
・ある位置(i,j)で手番が来た時に、
・どこか一か所でも相手が詰む位置（相手が動けなくて負けになる位置）に移動できれば勝ち
・どこに移動しても相手が勝つ位置であれば負け
となる。

また、
・壁や障害物の中にいる状態で手番が回ってくる -> 直前の手番で相手が詰んでいるので勝ち
と考える。
"""


H,W = map(int, input().split())
S = [input() for _ in range(H)]

win_first = [[None for _ in range(W)] for _ in range(H)]

# マス(i,j)で先攻の順番が来た時に勝つか負けるか
def solve(i,j):
    # 壁か障害物の中で順番が回ってくることはない（その前の手番は後攻が動かせない）ので先攻の勝ち
    if i >= H or j >= W or S[i][j] == "#":
        return True
    
    if win_first[i][j] != None:
        return win_first[i][j]

    ret = False
    for dx,dy in [(1,0), (0,1), (1,1)]:
        # 次の手番が後攻なので、その時に後攻が負けとなる移動先があるなら先攻の勝ち
        if not solve(i+dy, j+dx):
            ret = True

    win_first[i][j] = ret
    return ret

print("First" if solve(0, 0) else "Second")
```

```python
"""
三分探索

二分探索は単調増加や単調減少のときに使える
三分探索は、ある関数がある範囲で一つ極をもつときに使う。
http://kyopro.hateblo.jp/entry/2019/04/25/134128


下記が適当なフレームワーク
"""
hi = N
lo = 0
# 一回の計算で範囲が2/3なので、適当な回数回して極に収束させる
for _ in range(200):
    # 三分点の左側
    a = (2 * lo + hi) / 3.0
    # 三分点の右側
    b = (lo + 2 * hi) / 3.0

    # lo~a, a~b, b~hiの三つの区間ができる

    # a~hiの範囲で単調増加なので、最小値（関数の底）がlo~aにあり、つぎはlo~bを調べる
    if some_func(a) <= some_func(b):
        hi = b
    
    # a ~ bに底がある場合は、some_func(a)とsome_func(b)から単調増加か単調減少か決めることができないが、  
    # a~bは、loを切り上げてもhiを下ろしてきても、2/3にしたどちらの探索範囲にも含まれるので、
    # if 底がa~bにある場合～　のように場合分けする必要はなく、どちらかに含まれる
    else:
        lo = a

```

```python
"""

AtCoder Regular Contest 054  B - ムーアの法則

計算開始までの時間　＋　計算にかかる時間　を最小化

計測開始直後では計算時間はPかかるが、x待つと計算速度が2**(x/1.5)になる
ので、計測開始からの経過時間をxとして、
x + P / (2**(x/1.5)) の最小値を求めると良いか？

二分探索は単調増加とか単調減少の時に使うが、今回は違うので使えない。
（イメージ的には、今回の関数は歪んだUみたいな形になってるはず）

で、底の部分が欲しい（微分して傾き0のところ）が、微分はしたくない。
どこで単調増加していて、どこで単調減少してるかわかると嬉しい

三分探索やる


10**18と制限が大きいので、計算の途中でオーバーフローしたっぽい
割り算をやめてpowの指数を負にしておくと行けるか
"""



P = float(input())

if P == 0.0:
    print(P)
    exit()

def calc_time(x):
    return x + P * (pow(2.0, -x / 1.5))

hi = P
lo = 0
# 一回の計算で範囲が2/3なので、200も回せば十分か
for _ in range(200):
    # 三分点の左側
    a = (2 * lo + hi) / 3.0
    # 三分点の右側
    b = (lo + 2 * hi) / 3.0
    # 左が右より大きいので、右側にそこがある
    if calc_time(a) > calc_time(b):
        lo = a
    else:
        hi = b

# 下界の方
print(calc_time(lo))
```

```python
"""
フェルマーの小定理
https://ja.wikipedia.org/wiki/%E3%83%95%E3%82%A7%E3%83%AB%E3%83%9E%E3%83%BC%E3%81%AE%E5%B0%8F%E5%AE%9A%E7%90%86

aとpが互いに素でpが素数の時、
(a**(p-1)) % p = 1
-> (a * a**(p-2)) % p = 1
これより、aの逆元(a**(-1)) （mod p の時の）は、 a**(p-2)

(m / n) % (10**9+7) が知りたい時、例えば配列要素にこれを実施していくとTLEになることが多い
-> m / n = m * n**(-1)であり
上の逆元を用いると、aの逆元(a**(-1)のこと)は、 a**(p-2) になることから、
(m / n) % MOD = m * n**(-1) % MOD = m * n**(MOD - 2) % MOD   (競プロの問題的に、MOD = 10**9+7としておく)
が言える

累乗はたとえば15乗する場合に
13 = 8 + 4 + 1
みたいに2進数にしたときに１が立っている部分の和で求められるので、それなりに大きい数が指数についていても、割と高速に求められるので、割り算よりは指数にしたい
"""
```

```python
"""
AtCoder Beginner Contest 152  E - Flatten

Aの最小公倍数を求めてAの要素で割ったものを足した余りを求めるでpythonはギリギリACだが、定石として毎回MODをとるのが一般的（と思う）
が、そうするとTLEになってしまう。これはpypyも同様のはず。

割り算の計算が遅いので、フェルマーの小定理を利用する。
pow(値, 非常に大きな値)が遅くなるようにも思えるが、たとえば１３乗するような例を考えると、
13 = 8 + 4 + 1
のように、二進数に直したときに1が立っている部分の和として求められるので、a**4,a**8などが分かっていると高速に計算できる
"""

from collections import defaultdict


N = int(input())
A = list(map(int, input().split()))
MOD = 10**9 + 7

primes_total = defaultdict(int)

for a in A:
    tmp = a
    primes_a = defaultdict(int)
    p = 2
    while p * p <= tmp:
        while tmp % p == 0:
            primes_a[p] += 1
            tmp //= p
        p += 1
    if tmp != 1:
        primes_a[tmp] += 1

    for k in primes_a.keys():
        primes_total[k] = max(primes_total[k], primes_a[k])

lcm = 1
for k in primes_total.keys():
    lcm *= pow(k, primes_total[k], MOD)
    lcm %= MOD

ans = 0

for a in A:
    # フェルマーの小定理より
    ans += lcm * pow(a, MOD - 2, MOD)
    ans %= MOD

print(ans)
```

# XOR


## XORが出てきたら思い出したいこと
 - c1~cnに対してxorをとる ＝ それぞれcについて、各bitが立っている個数が奇数のbitに１が立ち、偶数のbitに０が立つ
   - なので、複数の数に対しxorをとるとき、順番を入れ替えても結果は同じ
   - bitごとに考えるといろいろ計算が楽になるかも
 - ある偶数nについて、 n xor (n+1) = 1  (下位１bit以外が同じなので)
 - a xor 0 = a
 - a xor a = 0
 - A,A+1,...,Bについてxorをとる = {0~A-1のxor} xor {0~Bのxor}
   - 0~A-1までが打ち消しあうため


```python


"""
AtCoder Beginner Contest 121  D - XOR World


・c1~cnについて全体のxorをとるとき、それぞれのcについて各bitで１が奇数個立っているとそのbitが１、偶数個立っているとそのbitが０になる
　→　xorは順番を入れ替えても答えは変わらない

・0 xor N = N、A xor~B = (0 xor ~A-1) xor (0 xor ~B)

・それぞれの偶数nについて、n xor (n+1) = 1 (下位１ビットだけが異なるから)
"""


A, B = map(int, input().split())

def solve(n):

    if n == 0:
        return 0

    nums = (n + 1) % 4

    if nums == 0:
        return 0

    if nums == 1:
        return n

    if nums == 2:
        return 1 # xが偶数で、X ^ (x+1) = 1 なので

    if nums == 3:
        return 1 ^ n


print( solve(A-1) ^ solve(B))
```

# 整数の問題
 - 配列A全体の最大公約数 = 要素一つずつについて最大公約数とっていく
 - 数字Sの部分文字列について
   - S\[l:r] = (S\[l:N] - S\[r:N]) / 10**x のように、下何桁か切り取って10のx乗で割ってやると間を切り取れる
   - S\[l:r]についてXXXで割り切れるか、みたいな場合は、XXXが10と互いに素であれば、S\[l:N],S\[r:N]のmod XXX が等しければ、S\[l:r]も割り切れるなど



```python
"""

AtCoder Beginner Contest 164  D - Multiple of 2019

何回か前のE問題で似たようなのが出ていた。

（１）S[l:r] = (S[l:] - S[r:]) / 10**xx
（２）Xと10が互いに素の場合、(S[l:] - S[r:])がXで割り切れると(S[l:] - S[r:])  10**xx　もXで割り切れる
（３）Xで割ってY余る数同士の差分をXで割ると余り0

を理解しているとよいかと。特に（１）を思いつかなければ話にならないのでこれはしっかり。

また、今回の制限だと、Sが長すぎるときに、毎回部分文字列をintにしてMODをとるとTLEになるはずなので、一番右をi=1として、
・右からi番目まで = 右からi番目の数字 * pow(10, i-1) + i-1番目までの合計
の形で考えて、S[l:]を考えられる。また、
・操作の都度そこまでをX（=2019）で割った余りとその個数の対応を求めておくことで、rを何個選びうるか
・r=0（引かない）を最初に考慮しておく
・pow(n,x,mod)の利用と、i番目までの数字をMODの状態でもっておく
により高速化することで十分間に合う
"""

from collections import defaultdict

S = input()
N = len(S)
d = defaultdict(int)

ans = 0
curr = 0
d[0] += 1 # S[l:] - S[r:]で、rの方に何も選ばない場合（S[l:]が2019で割りきれるかを考えるとき）
for i in range(N):
    curr = int(S[N-1-i]) * pow(10, i, 2019) + curr
    curr %= 2019
    ans += d[curr]
    d[curr] += 1

print(ans)
```

### bit関係の問題

```python




"""


第5回 ドワンゴからの挑戦状 予選　　B - Sum AND Subarrays

N<=1000より美しさの全列挙は可能
K個選んでbitごとの論理積の最大値

K個選ぶ方法はnCk(10**6?, K)とかなので、全探索は間に合わない

K個選んだもののうちの最大、を言い換えを考える。
bitごとの論理積をとるので、選ばれたK個の中で、bitが大きいものについて、K個すべてで1が立っていると嬉しい。
また、2**0 + ... + 2**(N-1) <= 2**N なので、大きいbitが一個立ってる方が、それより小さいもの全てが立っているより嬉しい。
→大きいbitから見ていって、各美しさでそのbitに1が立っているか調べて、K個以上立っていれば、その美しさの中からK個採用される。K個に満たなければ、引き続きその美しさ集合について、下位ビットが立っているか見る
　この時いずれの美しさも、今見ているbitが立っているので、答えにそのbit分加える。
　で、次のbitをみていって～を繰り返す。
　最下位bitまで見ても、美しさがK個以上のこっているなら、そこから適当にK個選んだら同じ答えになるので気にしなくてよい

bitごとの論理積が大きくなるようにK個選ぶ　→　大きいbitから、そのbitが立っているものがK個以上ある物が選ばれる
の言いかえ
"""

N,K = map(int, input().split())
A = list(map(int, input().split()))

accum = [0] * N
accum[0] = A[0]
for i in range(1,N):
    accum[i] = accum[i-1] + A[i]

beauty = []
for l in range(N):
    for r in range(l, N):
        if l == 0:
            beauty.append(accum[r])
        else:
            beauty.append(accum[r] - accum[l-1])


# 10**9 * 1000 から、だいたい2**40ぐらい
ans = 0
candidate = beauty
for i in range(40, -1, -1):
    nxt_candidate = []
    cnt = 0
    for b in candidate:
        if b & (1 << i):
            cnt += 1
            nxt_candidate.append(b)
    
    # iビット目が立っている個数がKに満たなければ、現在の美しさ集合について、より下位のビットを見る
    # K個以上あれば、iビット目が立ってない奴は選ばれないので除外できる
    if cnt >= K:
        ans += 2**i
        candidate = nxt_candidate

print(ans)
```
