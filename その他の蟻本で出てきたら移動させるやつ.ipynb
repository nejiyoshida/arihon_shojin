```python
"""
二次元累積和


普通の累積和を一番上の行でとって、それより下の行は、その行だけの累積和を変数に持ちつつ、一個上の行の累積和を足していく形
詳細
https://paiza.hatenablog.com/entry/2014/05/28/%E3%82%82%E3%81%97%E5%A5%B3%E5%AD%90%E5%A4%A7%E7%94%9F%E3%83%97%E3%83%AD%E3%82%B0%E3%83%A9%E3%83%9E%E3%81%AB%E3%80%8E%E3%82%A2%E3%83%AB%E3%82%B4%E3%83%AA%E3%82%BA%E3%83%A0%E3%80%8F%E3%82%92%E5%9B%B3
"""

# arr[][] があるとする
def make_2d_accum(arr, h, w):
    accum_2d = [[0 for _ in range(w)] for _ in range(h)]
    accum_2d[0][0] = arr[0][0]
    # 一行目なので、ここは一次元と同じ
    for j in range(1,w):
        accum_2d[0][j] = accum_2d[0][j-1] + arr[0][j]
    # 下の部分の累積和なので二次元
    for i in range(1, h):
        # i行目の累積和
        accum_row_i = 0
        for j in range(w):
            accum_row_i += arr[i][j]
            # 一行上までとi行目を足して２次元
            accum_2d[i][j] = accum_2d[i-1][j] + accum_row_i

    return accum_2d

# (a,b)～(c,d)の部分の累積和。全部同じ(a=b=c=d)だと一マス分
def calc_accum_rectangle(accum_2d, a, b, c, d):
    if a > c or b > d:
        return 0
    elif a == 0 and b == 0:
        return accum_2d[c][d]
    elif a == 0:
        return accum_2d[c][d] - accum_2d[c][b-1]
    elif b == 0:
        return accum_2d[c][d] - accum_2d[a-1][d]
    else:
        return accum_2d[c][d] - accum_2d[a-1][d] - accum_2d[c][b-1] + accum_2d[a-1][b-1]
```

```python
"""
AtCoder Regular Contest 025  B - チョコレート

いかにも２次元累積和をやってくれというような問題。
白と黒の一方を負にしておけば問題ない。
"""

# arr[][] があるとする
def make_2d_accum(arr, h, w):
    accum_2d = [[0 for _ in range(w)] for _ in range(h)]
    accum_2d[0][0] = arr[0][0]
    # 一行目なので、ここは一次元と同じ
    for j in range(1,w):
        accum_2d[0][j] = accum_2d[0][j-1] + arr[0][j]
    # 下の部分の累積和なので二次元
    for i in range(1, h):
        # i行目の累積和
        accum_row_i = 0
        for j in range(w):
            accum_row_i += arr[i][j]
            # 一行上までとi行目を足して２次元
            accum_2d[i][j] = accum_2d[i-1][j] + accum_row_i

    return accum_2d

# (a,b)～(c,d)の部分の累積和。全部同じ(a=b=c=d)だと一マス分
def calc_accum_rectangle(accum_2d, a, b, c, d):
    if a > c or b > d:
        return 0
    elif a == 0 and b == 0:
        return accum_2d[c][d]
    elif a == 0:
        return accum_2d[c][d] - accum_2d[c][b-1]
    elif b == 0:
        return accum_2d[c][d] - accum_2d[a-1][d]
    else:
        # ２項目と３項目の重複部分を足すのを忘れない
        return accum_2d[c][d] - accum_2d[a-1][d] - accum_2d[c][b-1] + accum_2d[a-1][b-1]


H,W = map(int, input().split())
C = [list(map(int, input().split())) for _ in range(H)]


for i in range(H):
    for j in range(W):
        if i % 2  != j % 2:
            C[i][j] *= -1

accum_2d = make_2d_accum(C, H, W)

ans = 0
for b in range(W):
    for a in range(H):
        for d in range(b, W):
            for c in range(a, H):
                # 指定の範囲の二次元累積和が0であれば、その面積を現時点の最大と比較
                if calc_accum_rectangle(accum_2d, a, b, c, d) == 0:
                    ans = max(ans, (c - a + 1)*(d - b + 1))

print(ans)
```

```python
"""
heapqの使い道

ダイクストラ法とかでコストの確定してるコストが小さいノードから探索する～みたいに使うことが多いが、ほかの使い道もある。
たとえば、要素数N（10**5とか）の配列に対して、クエリ(l,r)が与えられるとき、配列の[l:r]にたいして＋１とか（ほかの値でもいいけど）する場合を考える。
愚直にクエリを実行していくと、クエリの数 x 配列サイズ　の計算量となり、膨大な実行時間がかかる。

ここでheapqをうまいこと使うと、計算量を減らせる。
下の問題がその例普通の問題ならTLE
"""
```

```python
"""
AtCoder Beginner Contest 014  C - AtColor


愚直に実行すると、(10**6 + 1) * N　の計算量になり間に合わない
ある濃度について、購入したい人が何人いるか考えるとき、
・事前にaが小さい順に並べておいた購入情報から、今考えている濃度より購入範囲の下限が小さいものを取り出す
・その数の分だけ購入人数をプラスする
で、その濃度の購入人数が分かる。
また、ここまででとりだした購入情報について、次の濃度の購入人数に影響しない（bが次の濃度より小さい）場合もある。
それらをのぞくため、
・上で取り出した購入情報は、上限が小さい順にheapqに積んでおく
・次の濃度について確認する前に、無向になる情報をheapqから捨て、捨てた数だけ購入人数から引いておく
でOK
"""


import heapq


N = int(input())
q = []
for _ in range(N):
    a,b = map(int, input().split())
    heapq.heappush(q, (a,b))


q_effective = []
# 濃度ごとに確認
points = [0] * (10**6 + 1)

curr_point = 0
for i in range(10**6 + 1):
    # 現在見ている濃度以下の購入範囲の下限がある場合は、得点を足す
    while len(q) >= 1 and q[0][0] <= i:
        curr_point += 1
        a,b = heapq.heappop(q)
        heapq.heappush(q_effective, (b, a))

    points[i] = curr_point

    # 次のiに進んだ時に、ある購入範囲が次のiの人気に寄与しなくなる物があるなら除外する
    while len(q_effective) >= 1 and q_effective[0][0] <= i:
        curr_point -= 1
        _,_ = heapq.heappop(q_effective)

print(max(points))
```
